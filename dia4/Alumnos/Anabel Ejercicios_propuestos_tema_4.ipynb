### Ejercicio 1 
Visualizar la evolución no de los casos totales sino de los casos nuevos de España mediante un gráfico de líneas.

```python
#Importamos datos
datos = pd.read_csv('https://raw.githubusercontent.com/SerDiaz/Introduccion_programacion_en_python/main/dia4/owid-covid-data.csv') 

#extraemos los datos de españa
datos_españa = datos[datos['location'] == 'Spain']

#importamos la libreria 'matplotlib' y 'panda'
import matplotlib.pyplot as plt
import pandas as pd

#tamaño letra de los ejes
import matplotlib as mpl
mpl.rcParams['font.size'] = 18.

#posible error en los datos de "new cases"? nos aparece un valor negativo (?)
datos_españa.loc[datos_españa.new_cases < 0, "new_cases"] = 0

#Datos representados en los ejes
x = datos_españa['date']
y = datos_españa['new_cases']

# Hacer el recuadro más grande
fig = plt.figure(figsize=(15,10))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])


#total fechas de españa
n = len(datos_españa)

# Vamos a indicar solo tres datos del eje x
f_ini = 0
f_med = (n-1) /2
f_fin = n-1

# Le añadimos los tres elementos que vamos a mostrar.
ax.set_xticks([f_ini, f_med, f_fin])
ax.set_xticklabels(['enero','junio','diciembre'])

#Titulo de la grfica, y nombre de los ejes
ax.set_xlabel('Fechas') # nombre eje x 
ax.set_ylabel('Cantidad de casos') # nombre eje y
ax.set_title('Nuevos casos en España') # nombre gráfica

#Datos del eje Y
ax.set_yticks([0,10000,20000,30000,40000,50000])
ax.set_yticklabels(['0','10 mil','20 mil', '30 mil', '40 mil', '50 mil'])


#mostramos por pantalla 
plt.plot(x, y, lw=2)
plt.show()

#guardamos el grafico
#fig.savefig("total_es_mes.png")
```

### Ejercicio 4 

Ahora hacer una gráfica de queso para los casos totales sólo de America del Sur.

```python
#Importamos datos
import pandas as pd

#Cambiamos el formato de la fecha para trabajar con ellos
datos['date'] =  pd.to_datetime(datos['date'], format='%Y-%m-%d') # Transformamos formato datos de 'string' a 'timestamp', que es otra librería de tiempo

#Filtramos para que nos quede solo los datos de los paises latinos
datos_asur =datos[datos['continent'] == 'South America'] # Reasignamos de nuevos los datos de America del Sur, ahora con el formato

#Filtramos los datos para que solo queden los datos referidos al dia 30 de Nov de 2020
datos_asur = datos_asur[(datos_asur['date'].dt.year == 2020) & (datos_asur['date'].dt.month == 11) & (datos_asur['date'].dt.day == 30)]

datos_asur[(datos_asur['date'].dt.year == 2020) & (datos_asur['date'].dt.month == 11) & (datos_asur['date'].dt.day == 30)]
```

```python
# Gráficas de queso
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.size'] = 15.0

values = datos_asur.total_cases
colors = ['#295292', '#0D2A58', '#0A1B35', '#5D0505', '#7D1E1E','#A62222','#36145D', '#B2AB0A', '#E0D716', 'k','#9130FF','w']
labels = datos_asur.location
plt.pie(values, colors= colors, labels=labels)
fig = plt.gcf()
fig.set_size_inches(15,15) # Para cambiar el tamaño

plt.title('Reparto de casos totales en America Latina')
plt.show()
fig.savefig("reparto_queso.png") # Calidad resolucion
```

### Ejercicio 5
Queremos lo mismo de la gráfica de puntos que vimos pero queremos ver la evolución de casos totales con puntos de junio, julio y agosto

```python
#Almacenamos los datos que cumplan la condición: 
cond_mes = datos['date'].dt.month >= 6
cond_mes2 = datos['date'].dt.month <= 8

datos_intervalo = datos[(cond_mes) & (cond_mes2)]

#paises que vamos a representar en el grafico
datos_int_italia = datos_intervalo[datos_intervalo['location'] == 'Italy']
datos_int_alemania = datos_intervalo[datos_intervalo['location'] == 'Germany']
datos_int_francia = datos_intervalo[datos_intervalo['location'] == 'France']
datos_int_reinoUnido = datos_intervalo[datos_intervalo['location'] == 'United Kingdom']
datos_int_españa = datos_intervalo[datos_intervalo['location'] == 'Spain']


mpl.rcParams['font.size'] = 20.0


cas_mill_es = datos_int_españa['date']
test_mil_es = datos_int_españa['total_cases']

cas_mill_uk = datos_int_reinoUnido['date']
test_mil_uk = datos_int_reinoUnido['total_cases']

cas_mill_it = datos_int_italia['date']
test_mil_it = datos_int_italia['total_cases']

cas_mill_ger = datos_int_alemania['date']
test_mil_ger = datos_int_alemania['total_cases']

cas_mill_fr = datos_int_francia['date']
test_mil_fr = datos_int_francia['total_cases']


fig = plt.figure()
ax = fig.add_axes([1, 1, 2, 2]) # main axes

ax.set_xlabel('Fecha') # nombre eje x 
ax.set_ylabel('Casos Totales') # nombre eje y
ax.set_title('Casos Totales en los paises:') # nombre gráfica


ax.scatter(cas_mill_es, test_mil_es, lw=3,label = "España")
ax.scatter(cas_mill_ger, test_mil_ger, lw=3,label = "Alemania")
ax.scatter(cas_mill_it, test_mil_it, lw=3,label = "Italia")
ax.scatter(cas_mill_fr, test_mil_fr, lw=3,label = "Francia")
ax.scatter(cas_mill_uk, test_mil_uk, label = "Reino Unido")


plt.legend()
fig.savefig("casos_totales.png") # Calidad resolucion

plt.show()
```
